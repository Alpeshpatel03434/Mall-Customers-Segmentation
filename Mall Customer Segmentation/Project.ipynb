## Clustering - Univariate, Bivariate, Multivariate

```python
from sklearn.cluster import KMeans
```

### Univariate Clustering and If Assume number of clusters = 3

```python
clustering1 = KMeans(n_clusters=3)
clustering1.fit(df[['Annual Income (k$)']])
```

```python
clustering1.labels_
```

```python
df['Income Cluster'] = clustering1.labels_
df.head()
```

```python
df['Income Cluster'].value_counts()
```

```python
clustering1.inertia_
```

```python
intertia_scores=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df[['Annual Income (k$)']])
    intertia_scores.append(kmeans.inertia_)
```

```python
intertia_scores
```

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1,11),intertia_scores)
plt.xlabel('Number of clusters', fontsize = 14)
plt.ylabel('WCSS', fontsize = 14) 
plt.show()
```

```python
df.columns
```

```python
df.groupby('Income Cluster')['Age', 'Annual Income (k$)',
       'Spending Score (1-100)'].mean()
```

### Bivariate Clustering If Assume number of clusters=5

```python
clustering2 = KMeans(n_clusters=5)
clustering2.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
df['Spending and Income Cluster'] =clustering2.labels_
df.head()
```

```python
intertia_scores2 = []
for i in range(1,11):
    kmeans2=KMeans(n_clusters=i)
    kmeans2.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
    intertia_scores2.append(kmeans2.inertia_)
```

```python
intertia_scores2
```

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1,11),intertia_scores2)
plt.xlabel('Number of clusters', fontsize = 14)
plt.ylabel('WCSS', fontsize = 14) 
plt.show()
```

```python
centers =pd.DataFrame(clustering2.cluster_centers_)
centers.columns = ['x', 'y']
```

```python
plt.figure(figsize=(14, 10))
plt.scatter(x=centers['x'],y=centers['y'], s=500, c='black', marker='*')
sns.scatterplot(data=df, x ='Annual Income (k$)', y='Spending Score (1-100)', s = 100, hue='Spending and Income Cluster', palette='tab10')
plt.xlabel("Annual Income (k$)", fontsize=14)
plt.ylabel("Spending Score (1-100)", fontsize=14)
plt.savefig('clustering_bivaraiate.png')
```

```python
pd.crosstab(df['Spending and Income Cluster'],df['Gender'],normalize='index')
```

```python
df.groupby('Spending and Income Cluster')['Age', 'Annual Income (k$)',
       'Spending Score (1-100)'].mean()
```

### Mulivariate Clustering

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
```

```python
df.head()
```

```python
new_df = pd.get_dummies(df, drop_first=True)
new_df.head()
```

```python
new_df.columns
```

```python
new_df = new_df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)','Gender_Male']]
new_df.head()
```

```python
new_df = scale.fit_transform(new_df)

new_df = pd.DataFrame(scale.fit_transform(new_df))
new_df.head()
```

```python
intertia_scores3 = []
for i in range(1,11):
    kmeans3=KMeans(n_clusters=i)
    kmeans3.fit(new_df)
    intertia_scores3.append(kmeans3.inertia_)
```

```python
intertia_scores3
```

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1,11), intertia_scores3)
plt.xlabel('Number of clusters', fontsize = 14)
plt.ylabel('WCSS', fontsize = 14) 
plt.show()
```
